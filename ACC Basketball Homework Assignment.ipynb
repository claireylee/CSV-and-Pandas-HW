```python
import pandas as pd 

1. # reading the csv file
# loading csv file 
file_path = '/Users/clairelee/Downloads/acc_players-2324F.csv'
df = pd.read_csv(file_path, skiprows=1) 

# first 5 rows 
df.head()
```

```python
2. #basic analysis 

#calculate the total points scored by all players combined 
total_points = df['PTS'].sum()

#find the player who has played the most minutes (MP)
most_minutes_player = df[df['MP'] == df['MP'].max()]['Player'].values[0]

#identify the top 5 players in terms of total rebounds (TRB)
top_5_rebounders = df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)

total_points, most_minutes_player, top_5_rebounders
```

```python
3. #player filtering

#create dataframe containing only players who played more than 500 minutes 
players_over_500_minutes = df[df['MP'] > 500]

#from this filtered dataframe, determine the player with the highest total assists(AST)
highest_assists_player = players_over_500_minutes[players_over_500_minutes['AST'] == players_over_500_minutes['AST'].max()]['Player'].values[0]

#who are the top 3 assist leads in the league?
top_3_assist_leaders = players_over_500_minutes[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)

#who are the top 3 shot blockers 
top_3_shot_blockers = players_over_500_minutes[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
highest_assists_player, top_3_assist_leaders, top_3_shot_blockers
```

```python
4. #school-based analysis 

#group the players by school and calculate the total points scored by each school 
school_total_points = df.groupby('School')['PTS'].sum().reset_index()

#group the players by school and calculate the total assists(AST) for each team 
school_total_assists = df.groupby('School')['AST'].sum().reset_index()
#sort the school by total points scored and display the top 3 schools 
top_3_schools_by_points = school_total_points.sort_values(by='PTS', ascending=False).head(3)
school_total_points, school_total_assists, top_3_schools_by_points
```
